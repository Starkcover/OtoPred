## Export checkpoints for Streamlit deployment
Run after training. Downloads a zip ready to put in the `checkpoints/` folder.


```python
import os, torch, numpy as np, joblib, zipfile

CKPT_DIR = '/kaggle/working/checkpoints'

# Save ensemble config
torch.save({'best_w': float(BEST_W), 'threshold': float(OPTIMAL_THRESHOLD)},
           f'{CKPT_DIR}/ensemble_config.pt')

# Save train_idx
np.save(f'{CKPT_DIR}/train_idx.npy', train_idx)

files = [
    'best_model_final.pt',
    'rf_baseline.pkl',
    'hetero_data_final.pt',
    'df_with_chembl.parquet',
    'protein_features_v2.npy',
    'protein_map_v2.parquet',
    'ensemble_config.pt',
    'train_idx.npy',
]

print("Checking files:")
for f in files:
    p = f'{CKPT_DIR}/{f}'
    exists = os.path.exists(p)
    size   = f'{os.path.getsize(p)/1e6:.1f} MB' if exists else '—'
    print(f"  {'OK' if exists else 'MISSING'}  {f}  ({size})")

zip_path = '/kaggle/working/ototox_checkpoints.zip'
with zipfile.ZipFile(zip_path, 'w', zipfile.ZIP_DEFLATED) as zf:
    for f in files:
        p = f'{CKPT_DIR}/{f}'
        if os.path.exists(p):
            zf.write(p, f'checkpoints/{f}')

size_mb = os.path.getsize(zip_path)/1e6
print(f"\nZip: {zip_path}  ({size_mb:.1f} MB)")
print("Download from Kaggle output panel.")
print("Extract into your streamlit app folder (creates checkpoints/ subfolder).")
```
